### Pivot is used to reshape or transform the data

```python
df.pivot(index = 'date',columns = 'city')
```

```python
df.pivot(index = 'date',columns = 'city',values = 'humidity')
```

```python
df.pivot(index = 'humidity',columns = 'city')
```

### Pivot Table is used to summarize and aggregate data inside dataframe

```python
df = pd.read_csv('weather.csv')
df
```

### Default aggregation is mean (but here each value is unique, so mean = actual value)

```python
df.pivot_table(index = "date", columns = "city")
```

```python
df.pivot_table(index = 'city',columns = 'date',aggfunc="sum")
```

```python
df.pivot_table(index = 'city',columns = 'date',aggfunc="diff")
#without assigning why this is not producing the output
```
