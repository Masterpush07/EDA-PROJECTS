### checking with duplicate values 

```python
df = pd.DataFrame({
    'Product':['A','B','C','A','B','C','C'],#C acts as a duplicate
    'Month':['JAN','JAN','JAN','FEB','FEB','FEB','FEB'], #Feb acts as a Duplicate
    'Qunatity_sold':[100,120,95,110,130,100,110],
    'Revenue':[5000,6000,4500,5500,7000,4800,5000]
})
```

```python
df
```

### here it will take the mean value (aggregation) whenever there is a duplicate

```python
df.pivot_table(index = 'Product',columns = 'Month')
```

```python
df.pivot_table(index = 'Product',columns = 'Month',aggfunc= sum )
```

```python
df.pivot_table(index = 'Product',columns = 'Month',aggfunc= 'sum', margins = True)   
```
