# importing the required libraries

```python
import pandas as pd
```

```python
df=pd.read_csv('movies.csv')
```

```python
df.info()
df.duplicated().any()
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].mode()[0])
df['Rotten Tomatoes']=df['Rotten Tomatoes'].fillna(df['Rotten Tomatoes'].mode()[0])
df['IMDb']=df['IMDb'].fillna(df['IMDb'].mean()).round(1)
df['Runtime']=df['Runtime'].fillna(df['Runtime']).mean().round(1)

for i in ['Directors','Genres','Country','Language']:
    df[i]=df[i].fillna('Unknown')
df.info()

df['Rotten Tomatoes']=(df['Rotten Tomatoes'].str.strip('%').astype(float)/10).round(1)


```

```python
df
```

# Data transfomation / Feature engineering

```python
plts_df=pd.melt(df,id_vars=['Title','IMDb','Rotten Tomatoes'],value_vars=['Netflix',	'Hulu',	'Prime Video',	'Disney+'],var_name='platforms',value_name='Availability')
```

```python
plts_df=plts_df.loc[plts_df['Availability']==1]
```

```python
plts_df=plts_df.drop(columns=['Availability'])
```

```python
df['tot_plt']=df[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum(axis=1)
```
