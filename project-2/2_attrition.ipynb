```python
hire_cnt=(df['hire_date'].dt.year).value_counts()
term_cnt=df.loc[df['status']=='Terminated','term_date'].dt.year.value_counts()


att_ret_cnt=pd.merge(hire_cnt,term_cnt,how='outer',left_index=True,right_index=True)
att_ret_cnt.columns=['hired_cnt','attr_cnt']
#print(att_ret_cnt.rename(columns={'count_x':'hired_cnt','count_y':'attr_cnt'}))
att_ret_cnt['attr_cnt']=att_ret_cnt['attr_cnt'].fillna(0)
att_ret_cnt['hired_cnt']=att_ret_cnt['hired_cnt'].fillna(0)

att_ret_cnt['ret_cnt']=(att_ret_cnt['hired_cnt']-att_ret_cnt['attr_cnt'])
print(att_ret_cnt)
df
```

# Analysis of Attrition and Retention Using Plots

 Q1) Identify the Top 3 Years with the Highest Attrition Rate
 - Observe the termination type- Observe the termination reason
 - Identify which job roles were affected
 - Identify which departments were affected
 - Identify which experience levels were affected


```python
att_ret_cnt['attr_cnt'].sort_values().tail(3)
```

```python
ord=att_ret_cnt['attr_cnt'].sort_values(ascending=False)

# sns.barplot(att_ret_cnt,y='attr_cnt',x=att_ret_cnt.index,order=ord.head(3).index)
a=sns.barplot(att_ret_cnt,y='attr_cnt',x=att_ret_cnt.index,order=ord.index[:3])
plt.xticks(rotation=90)
a.bar_label(a.containers[0])
plt.title('Top 3 Highest Attrition Rate')
plt.show()
```

```python
df['term_year']=df['term_date'].dt.year

top3=att_ret_cnt['attr_cnt'].sort_values(ascending=False).index[:3]
new_df=df.loc[(df['status']=='Terminated')&(df['term_year'].isin(top3))]
new_df
```

```python
#'Observation over Term Type and Term reason'
fg=sns.FacetGrid(new_df,col='term_year')

fg.map_dataframe(sns.countplot,x='term_reason',hue='term_type',palette='Set1')

plt.legend()
```

```python
#Identify which job roles were affected
fg=sns.FacetGrid(new_df,col='term_year')

fg.map_dataframe(sns.countplot,x='job', hue='term_reason',order=new_df['job'].value_counts().head(10).index,palette='Set1')
plt.legend(labels=['resignation','retirement','layoff'])


fg.set_xticklabels(rotation=90)
#new_df['job'].value_counts() #fro understanding
```

```python
#- Identify which departments were affected
fg=sns.FacetGrid(new_df,col='term_year')
fg.map_dataframe(sns.countplot,x='dname',order=new_df['dname'].value_counts().head().index)
fg.set_xticklabels(rotation=90)
```

```python
fg=sns.FacetGrid(new_df,col='term_year')
fg.map_dataframe(sns.countplot,x='Exp_cat',order=new_df['Exp_cat'].value_counts().head(10).index)
fg.set_xticklabels(rotation=90)
```
