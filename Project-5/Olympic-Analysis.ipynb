```python
def fetch_medal_tally(df, year, country):
    medal_df = df.drop_duplicates(subset=['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event', 'Medal'])
    flag = 0
    if year == 'Overall' and country == 'Overall':
        temp_df = medal_df
    if year == 'Overall' and country != 'Overall':
        flag = 1
        temp_df = medal_df[medal_df['region'] == country]
    if year != 'Overall' and country == 'Overall':
        temp_df = medal_df[medal_df['Year'] == int(year)]
    if year != 'Overall' and country != 'Overall':
        temp_df = medal_df[(medal_df['Year'] == year) & (medal_df['region'] == country)]

    if flag == 1:
        x = temp_df.groupby('Year').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Year').reset_index()
    else:
        x = temp_df.groupby('region').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold',ascending=True).reset_index()

    x['total'] = x['Gold'] + x['Silver'] + x['Bronze']
    print(x)
```

```python
def most_successful(df, sport):
    temp_df = df.dropna(subset=['Medal'])

    if sport != 'Overall':
        temp_df = temp_df[temp_df['Sport'] == sport]

    # Count medals by athlete
    top_athletes = (temp_df['Name']
                    .value_counts()
                    .reset_index()
                    .head(15))
    
    # Rename columns for clarity
    top_athletes.columns = ['Name', 'Medal_Count']

    # Merge with original df to get Sport and Region
    x = (top_athletes
              .merge(df, on='Name', how='left')
              [['Name', 'Medal_Count', 'Sport', 'region']]
              .drop_duplicates('Name'))

    return x
```

## Exploratory Data Analysis(visuals)

#### 2.  Average Age of Athletes by Sex

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Sex', y='Age', data=df)
plt.title('Age Distribution by Gender')
plt.show()
print("Observation: Age distribution is similar, mostly 20–30 years old.")
```

#### 3. Top 10 Countries by Medal Count

```python
medals = df[df['Medal'].notna()].groupby('Team')['Medal'].count().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='Medal', y='Team', data=medals)
plt.title('Top 10 Countries by Medals')
plt.show()
print("Observation: USA, Russia, Germany, and China dominate medal counts.")
```

#### 4. Average Age per Sport

```python
sport_age = df.groupby('Sport')['Age'].mean().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='Age', y='Sport', data=sport_age)
plt.title('Top 10 Sports by Average Age of Athletes')
plt.show()
print("Observation: Older athletes in sports like Sailing, Shooting, Equestrian.")
```

#### 5. Height vs Weight Scatter Plot

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Height', y='Weight', hue='Sex', data=df)
plt.title('Height vs Weight of Athletes')
plt.show()
print("Observation: Males generally taller and heavier; some sports have outliers.")
```

#### 6.  Medal Distribution by Gender

```python
plt.figure(figsize=(6,6))
df[df['Medal'].notna()]['Sex'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue','lightgreen'])
plt.title('Medal Distribution by Gender')
plt.ylabel('')
plt.show()
print("Observation: Male athletes have historically won more medals, but gap is reducing.")
```

#### 7. Athletes per Olympic Year


```python
plt.figure(figsize=(10,5))
sns.countplot(x='Year', data=df, palette='viridis')
plt.title('Number of Athletes per Olympic Year')
plt.xticks(rotation=90)
plt.show()
print("Observation: Number of athletes increases over time; Winter Olympics have fewer athletes.")
```

#### 8.Top 10 Multi-Medal Athletes

```python
multi_medal = df[df['Medal'].notna()].groupby('Name')['Medal'].count().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='Medal', y='Name', data=multi_medal)
plt.title('Top 10 Athletes with Most Medals')
plt.show()
print("Observation: Some athletes excelled across multiple events or editions.")
```

#### 9.  Medal Count per Season (Summer vs Winter)

```python
season_medals = df[df['Medal'].notna()].groupby('Season')['Medal'].count().reset_index()
plt.figure(figsize=(6,4))
sns.barplot(x='Season', y='Medal', data=season_medals)
plt.title('Medals per Season')
plt.show()
print("Observation: Summer Olympics have far more medals than Winter Olympics.")
```

## Conclusion

Participation Trends: Male athletes historically outnumber females, but female participation has steadily increased over the years.

Age & Physical Attributes: Most Olympians are young adults (20–30 years); height and weight vary widely depending on the sport.

Country Dominance: USA, Russia, Germany, and China lead in medal counts, showing strong sports infrastructure and investment.

Medal Insights: Summer Olympics dominate medal counts compared to Winter Olympics, and some athletes have won multiple medals across events and editions.

Gender & Medals: Male athletes have historically won more medals, but female athletes are closing the gap in recent years.

Trend Over Time: Number of athletes per Olympic year has generally increased, reflecting the global growth of the Games.
