##### Check for Duplicate Values
##### Replace with Mean,Mode,Median
##### astype(),to_datetime()

```python
#there are no duplicate values
tit.duplicated().any() 
```

```python
tit.info()
```

```python
age_mean = tit['age'].mean()
fare_mean=tit['fare'].mean()
emb_mode=tit['embarked'].mode()[0]
tit['age']=tit['age'].fillna(age_mean)   #replacing null values with mean
tit['fare']=tit['fare'].fillna(fare_mean)  #replacing null values with mean
tit['embarked']=tit['embarked'].fillna(emb_mode)  #replacing null values with mode
tit=tit.drop(columns=['cabin'])
tit.info()
```

# 5) Data Transformation / Feature Engineering


```python
tit['age'] = tit['age'].round(1)
tit['fare'] = tit['fare'].round(1)
tit.head(3)
```

```python
#renaming 
tit = tit.rename(columns = {'sex':'Gender'})
tit.head()
```

```python
#title case that is first letter capital and remaining lower case
tit['Gender']=tit['Gender'].str.title()
tit
```
